```python
import selenium
from selenium import webdriver
import time
from selenium.webdriver.common.by import By

import pandas as pd
```

```python
df = pd.read_csv('vnexpress_categories.csv')

titles = df['title'].tolist()
links = df['links'].tolist()

PATH = r"D:\misc\chromedriver-win64\chromedriver.exe"

driver = webdriver.Chrome(PATH)
driver.maximize_window()
count = 0

```

```python
news_title = []
news_url = []
```

```python
for idx, url in enumerate(links):
    driver.get(url)
    time.sleep(3)
    while True:
        try:
            news = driver.find_elements(By.CSS_SELECTOR, 'h3.title-news a')
            print(f'Category: {titles[idx]}')
            for n in news:
                if len(n.text) < 10:
                    continue
                news_title.append(n.text)
                news_url.append(n.get_attribute('href'))
                count += 1
            next_page = driver.find_element(By.XPATH, '//*[@id="pagination"]/div/a[5]')
            driver.execute_script("arguments[0].click();", next_page)
            if count > 300:
                count = 0
                break
        except:
            break
    print('_____________________________________________________________________')
    
```

```python
df_1 = pd.DataFrame({'title': news_title, 'url': news_url})
df_1.to_csv('vnexpress_news.csv', index=False)

```
