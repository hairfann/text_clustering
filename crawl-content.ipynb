```python
import selenium
from selenium import webdriver
import time
from selenium.webdriver.common.by import By

import pandas as pd
import json
```

```python
# Load JSON file into a dataframe
with open(r'D:\references\DEADLINE 2nd Mar\vnexpress_news.json') as f:
    data = json.load(f)
    df = pd.DataFrame(data)
```

```python
df
```

```python
news_url = df['url'].tolist()
news_title = df['title'].tolist()
content = df['content'].tolist()
PATH = r"D:\misc\chromedriver-win64\chromedriver.exe"
```

```python
PATH = r"D:\misc\chromedriver-win64\chromedriver.exe"

driver = webdriver.Chrome(PATH)
driver.maximize_window()
count = 0
tmp = ''

for e, url in enumerate(news_url):
    try:
        if content[e] != 'Error':
            continue
        driver.get(url)
        time.sleep(3.14)
        articles = driver.find_elements(By.CLASS_NAME, 'Normal')    
        for a in articles:
            tmp += a.text + '\n'
        content[e] = str(tmp)
        print(f'Finished no.{e+1} article')
        tmp = ''
    except:
        print(f'Error at {e+1}')
        content[e] = 'Error'
        continue
df['content'] = content
```

```python
with open('vnexpress_news.json', 'w', encoding='utf-8') as f:
    df[['title', 'url', 'content']].to_json('vnexpress_news.json', orient='records', force_ascii=False)
        
```
