```python
import pandas as pd
import json 

# Load data
with open(r'D:\references\DEADLINE 2nd Mar\vnexpress_news.json') as f:
    data = json.load(f)
    df = pd.DataFrame(data)

```

```python
df
```

```python
df['segmented'] = df['content'].apply(lambda x: rdrsegmenter.word_segment(x))
```

```python
df['content'] = df['content'].apply(lambda x: x.replace('\n', ' '))
```

```python
with open('vnexpress_news.json', 'w', encoding='utf-8') as f:
    df[['title', 'url', 'segmented']].to_json('vnexpress_news_segmented.json', orient='records', force_ascii=False)
        
```
